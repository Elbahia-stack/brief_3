```python
import tensorflow as tf
import numpy as np
import pandas as pd


fashion_minist=tf.keras.datasets.fashion_mnist
(X_train,y_train),(X_test,y_test)=fashion_minist.load_data()

print("x_train",X_train.shape)
print("y_train",y_train.shape)
print("x_test",X_test.shape)
print("class de dataset",np.unique(y_train))

```

```python
import matplotlib.pyplot as plt
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i], cmap="Pastel1")
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Dense, Flatten
X_train=X_train/255
X_test=X_test/255


callback =tf.keras.callbacks.EarlyStopping(monitor='loss',
                                              patience=3)
model1= Sequential([
    Input(shape=(28, 28)),
    Flatten(),
    Dense(128, activation='relu',name="entree"),
    Dense(64, activation='relu',name="hidding"),
    Dense(10, activation='softmax',name="output")])
model1.summary()
model1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history=model1.fit(X_train, y_train, epochs=10 ,validation_data=(X_test,y_test))
```

```python
test_loss, test_accuracy = model1.evaluate(X_test, y_test)
```

### Documentation & Synthèse

#### Hyperparamètres testés
- Optimizer : `adam`
- Couches cachées : 128 et 64 neurones
- Activation : `relu`
- Epochs : 10
- Callback : `EarlyStopping(monitor='val_loss', patience=3)`

#### Choix de la fonction de perte
Nous avons utilisé `sparse_categorical_crossentropy` car les labels du dataset sont des entiers (0 à 9).  
Cette fonction est équivalente à `categorical_crossentropy`, mais évite d’avoir à encoder les labels en one-hot vectors.  
Elle est donc plus simple et plus rapide pour ce type de données.
